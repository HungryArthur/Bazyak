# **1.2 Математические Задачи**
Даны координаты трех вершин треугольника (х1, y1), (x2, y2),
(x3, y3). Найти его периметр и площадь.

```python
x1, y1 = map(float, input("Введите координату 1 вершины треугольника").split())
x2, y2 = map(float, input("Введите координату 2 вершины треугольника").split())
x3, y3 = map(float, input("Введите координату 3 вершины треугольника").split())

AB = ((x2 - x1) ** 2 + (y2 - y1) ** 2) ** 0.5
BC = ((x3 - x2) ** 2 + (y3 - y2) ** 2) ** 0.5
CA = ((x1 - x3) ** 2 + (y1 - y3) ** 2) ** 0.5

print("Периметр:", AB + BC + CA) # Перимитр
print("Площадь:", 1/2 * abs(x1 * (y2 - y1) + x2 * (y3 - y2) + x3 * (y1 - y3))) # Площадь
```

# **1.3 Логические выражения**
Составить программы, печатающие значение True, если следующие указанные высказывания являются истинными, и значение False - в противном случае.


Сумма цифр заданного трехзначного числа n является четным ЧИСЛОМ.

```python
num = int(input())
a, b, c = num // 100, num // 10 % 10, num % 10
sm = a + b + c # Сумма цифр
print(sm % 2 == 0)
```

# **1.4 Области, описываемые логическими выражениями**

Составить программу, которая печатает True, если точка с координатами (х, у) принадлежит заданным закрашенным (заштрихованным) областям, показанным на рисунках в табл. 1, и False - в противном случае.

```python
x, y = map(float, input().split())

print((x >= 0) and (9 <= (x**2 + y**2) <= 25))
```
